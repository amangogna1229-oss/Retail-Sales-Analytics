## Trim Extra Spaces

```python
from pyspark.sql.functions import trim, col

customers = customers.withColumn(
    "customer_id",
    trim(col("customer_id"))
)
```

```python
customers = customers.withColumn(
    "customer_name",
    trim(col("customer_name"))
)
```

```python
location = location.withColumn(
    "postal_code",
    trim(col("postal_code"))
).withColumn(
    "city",
    trim(col("city"))
).withColumn(
    "state",
    trim(col("state"))
).withColumn(
    "region",
    trim(col("region"))
).withColumn(
    "country_region",
    trim(col("country_region"))
)
```

```python
orders = orders.withColumn("order_id", trim(col("order_id"))) \
               .withColumn("ship_mode", trim(col("ship_mode"))) \
               .withColumn("customer_id", trim(col("customer_id"))) \
               .withColumn("product_id", trim(col("product_id")))
```

```python
products = products.withColumn("product_id", trim(col("product_id"))) \
                   .withColumn("category", trim(col("category"))) \
                   .withColumn("sub_category", trim(col("sub_category"))) \
                   .withColumn("product_name", trim(col("product_name")))
```

```python
from pyspark.sql.functions import regexp_replace, col

orders = orders.withColumn(
    "sales",
    regexp_replace(col("sales"), ",", ".")
)

orders = orders.withColumn(
    "discount",
    regexp_replace(col("discount"), ",", ".")
)

orders = orders.withColumn(
    "profit",
    regexp_replace(col("profit"), ",", ".")
)
```

```python
from pyspark.sql.types import DoubleType

orders = orders.withColumn(
    "sales",
    col("sales").cast(DoubleType())
)

orders = orders.withColumn(
    "discount",
    col("discount").cast(DoubleType())
)

orders = orders.withColumn(
    "profit",
    col("profit").cast(DoubleType())
)
```

```python
orders.printSchema()
```

## Saving Silver Tables

```python
customers.write.mode("overwrite").format("delta").saveAsTable("silver_customers")

orders.write.mode("overwrite").format("delta").saveAsTable("silver_orders")

products.write.mode("overwrite").format("delta").saveAsTable("silver_products")

location.write.mode("overwrite").format("delta").saveAsTable("silver_location")
```
