```python
customers = spark.read.format("csv") \
    .option("header", "true") \
    .option("sep", ";") \
    .option("inferSchema", "true") \
    .load("Files/RawData/Customers.csv")

orders = spark.read.format("csv") \
    .option("header", "true") \
    .option("sep", ";") \
    .option("inferSchema", "true") \
    .load("Files/RawData/Orders.csv")

products = spark.read.format("csv") \
    .option("header", "true") \
    .option("sep", ";") \
    .option("inferSchema", "true") \
    .load("Files/RawData/Products.csv")

location = spark.read.format("csv") \
    .option("header", "true") \
    .option("sep", ";") \
    .option("inferSchema", "true") \
    .load("Files/RawData/Location.csv")
```

```python
customers = customers.toDF(
    *[
        c.strip()
         .lower()
         .replace(" ", "_")
         .replace("/", "_")
         .replace("-", "_")
        for c in customers.columns
    ]
)
```

```python
orders = orders.toDF(
    *[
        c.strip()
         .lower()
         .replace(" ", "_")
         .replace("/", "_")
         .replace("-", "_")
        for c in orders.columns
    ]
)
```

```python
products = products.toDF(
    *[
        c.strip()
         .lower()
         .replace(" ", "_")
         .replace("/", "_")
         .replace("-", "_")
        for c in products.columns
    ]
)
```

```python
location = location.toDF(
    *[
        c.strip()
         .lower()
         .replace(" ", "_")
         .replace("/", "_")
         .replace("-", "_")
        for c in location.columns
    ]
)
```

```python
customers.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("bronze_customers")

orders.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("bronze_orders")

products.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("bronze_products")

location.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("bronze_location")
```
