```python
customers = spark.read.table("silver_customers")

orders = spark.read.table("silver_orders")

products = spark.read.table("silver_products")

location = spark.read.table("silver_location")
```

```python
gold_df = orders \
.join(customers, "customer_id", "left") \
.join(products, "product_id", "left") \
.join(location, "postal_code", "left")
display(gold_df)
```

```python
gold_df.write \
.mode("overwrite") \
.format("delta") \
.saveAsTable("gold_retail")
```
